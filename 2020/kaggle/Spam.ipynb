```python
morph = pymorphy2.MorphAnalyzer()
morph.parse('секси')
```

```python
def word_cut_tittle(text):
    uptext = ''
    text = text.replace('_', ' ')
    for word in tokenize(text):
        morph_word = morph.parse(word.text)[0]
        if any(['PREP' in morph_word.tag, 
                'CONJ' in morph_word.tag, 
                'INTJ' in morph_word.tag, 
                'PNCT' in morph_word.tag,
                'UNKN' in morph_word.tag,
                'NUMB' in morph_word.tag,
                '�' in word.text]):
            continue
        split_w = 'porn'
        if split_w in word.text:
            uptext += f'порно '
            continue
        split_w = 'sex'
        if split_w in word.text:
            uptext += f'секс '
            continue
        split_w = 'секс'
        if split_w in word.text:
            uptext += f'секс '
            continue
        split_w = 'оргия'
        if split_w in word.text:
            uptext += f'секс '
            continue
        split_w = 'intim'
        if split_w in word.text:
            uptext += f'секс '
            continue
        split_w = 'fuck'
        if split_w in word.text:
            uptext += f'секс '
            continue
        split_w = 'соитие'
        if split_w in word.text:
            uptext += f'секс '
            continue
        uptext += f'{morph_word.normal_form} '
    return uptext[:-1]

def word_cut_url(text):
    uptext = ''
#     text = text[:text.rfind('.')]
    for word in tokenize(text):
        morph_word = morph.parse(word.text)[0]
        if any(['PREP' in morph_word.tag, 
                'CONJ' in morph_word.tag, 
                'INTJ' in morph_word.tag, 
                'PNCT' in morph_word.tag,
                'NUMB' in morph_word.tag,
                'www' in word.text,
                '�' in word.text]):
            continue
        split_w = 'porn'
        if split_w in word.text:
            uptext += f'{split_w} '
            continue
        split_w = 'sex'
        if split_w in word.text:
            uptext += f'{split_w} '
            continue
        split_w = 'intim'
        if split_w in word.text:
            uptext += f'sex '
            continue
        uptext += f'{morph_word.normal_form} '
    return uptext[:-1]
```

```python
data.title = data.title.map(word_cut_tittle)
```

```python
data.url = data.url.map(word_cut_url)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer_title = TfidfVectorizer()
X_title = vectorizer_title.fit_transform(data.title)

vectorizer_url = TfidfVectorizer()
X_url = vectorizer_url.fit_transform(data.url)
```

```python
data
```

```python
from scipy.sparse import coo_matrix, hstack
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

X = hstack([X_title, X_url]) # , data.length.values.reshape((-1,1))
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.009, random_state=1234)
# rs=42 t_s=0.1
```

```python
lin_model = LogisticRegression(
                           class_weight={0: 1.5, 1: 10}, 
                           C=15,
                           warm_start=False, 
                           max_iter=3000, 
#                            l1_ratio=2,
                           tol=0.0001,
#                            solver='l',
#                            intercept_scaling=1,
                           random_state=1234)
```

```python
lin_model.fit(X_train, y_train)
```

```python
res = lin_model.predict(X_test)
f1_score(res, y_test)
```

```python
badres = data.loc[y_test.index]
badres['result'] = res
badres.loc[list(y_test[y_test != res].index)]
```
